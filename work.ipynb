```python
# XÁC ĐỊNH TAM GIÁC
a,b,c=map(float, input().split())

if a+b>c and a+c>b and b+c>a:
  if a==b==c:
   print("Tam giác đều")
  elif (a==b and a!=c) or (a==c and a!=b) or (b==c and b!=a):
   print("Tam giác cân")
  else:
    print("Tam giác thường")
else:
   print("Không phải tam giác")
```
