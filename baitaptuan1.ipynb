```python
# Bài 2
a = float(input("Nhập chiều rộng (a): "))
b = float(input("Nhập chiều dài (b): "))
r = float(input("Nhập bán kính khu vui chơi (r): "))

S_dat = a * b                
S_tron = 3.14 * r ** 2       
S_conlai = S_dat - S_tron   

print("Phần diện tích đất còn lại là:", round(S_conlai, 2))
```

```python
# Bài 3
c = input("Nhập 1 ký tự: ")

if c.isupper():                      # Nếu là chữ in hoa
    print("Chữ thường tương ứng là:", c.lower())
elif c.islower():                    # Nếu là chữ thường
    print("Chữ hoa tương ứng là:", c.upper())
else:
    print("Ký tự không phải là chữ cái.")
```

```python
# Bài 4
c = input("Nhập 1 ký tự: ")

if c.isalpha():       # isalpha() kiểm tra có phải là chữ cái (A-Z, a-z)
    print(f"{c} là ký tự alphabet.")
else:
    print(f"{c} không phải là ký tự alphabet.")
```

```python
# Bài 5
A = input("Nhập vào một chữ cái hoa (A-Z): ")

if A == 'A':
    print("Chữ cái liền trước là: Z")
else:
    truoc = chr(ord(A) - 1)  # Lùi lại 1 ký tự trong bảng mã ASCII
    print("Chữ cái liền trước là:", truoc)
```

```python
# Bài 6
a = float(input("Nhập cạnh a: "))
b = float(input("Nhập cạnh b: "))
c = float(input("Nhập cạnh c: "))

if a + b > c and a + c > b and b + c > a:
    p = (a + b + c) / 2
    S = (p * (p - a) * (p - b) * (p - c)) ** 0.5
    print("Diện tích tam giác là:", round(S, 3))
else:
    print("Ba cạnh này không tạo thành tam giác.")
```

```python
# Bài 8
ten = input("Tên chủ hộ: ")
cu = int(input("Chỉ số tháng trước: "))
moi = int(input("Chỉ số tháng này: "))

dien = moi - cu

if dien <= 50:
    tien = dien * 1984
elif dien <= 100:
    tien = 50 * 1984 + (dien - 50) * 2050
elif dien <= 200:
    tien = 50 * 1984 + 50 * 2050 + (dien - 100) * 2830
elif dien <= 300:
    tien = 50 * 1984 + 50 * 2050 + 100 * 2830 + (dien - 200) * 2997
elif dien <= 400:
    tien = 50 * 1984 + 50 * 2050 + 100 * 2830 + 100 * 2997 + (dien - 300) * 3350
else:
    tien = (50 * 1984 + 50 * 2050 + 100 * 2830 + 100 * 2997 +
            100 * 3350 + (dien - 400) * 3460)

# Tính VAT 8%
tong = round(tien * 1.08)

# In kết quả
print(f"Họ và tên: {ten}")
print(f"Tiền phải trả: {tong} đồng")
```
